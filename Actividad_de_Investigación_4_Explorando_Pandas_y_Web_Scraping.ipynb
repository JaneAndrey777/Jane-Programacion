```python
import pandas as pd
#Proceso de web scrapping a el BCCR
df_web  = pd.read_html('https://gee.bccr.fi.cr/IndicadoresEconomicos/Cuadros/frmConsultaTCVentanilla.aspx', decimal= ',',thousands='.', encoding='utf-8')
```

```python

#Inspecci贸n contenido del DataFrame
df = df_web[2].copy() #Copia del indice 2 de la lista de dataframe cargados
#df_web.clear()


#Obteber toda la informaci贸n del DataFrame
print('Tipo de estructura:')
print(type(df))


#Ver estructura
print('\n --------------- Dimensiones de Data Frame ---------------')
print(df.shape)
print('\n --------------- Informac贸n de Data Frame ---------------')
print(df.info())
print('\n --------------- Encabezado de Data Frame ---------------')
df.head()
```

## Secci贸n 2: Trabajamos en la limpieza de datos


* Eliminar los nulos
* Reparar el encabezado
* Arreglar tipos de datos

```python
#Reparar los encabezados

#1. Extraer la primera file como encabeza
encabezados = df.iloc[0]
encabezados

#2 Asignar la fila al encabezado de columnas DF
df.columns = encabezados

#3 Eliminar la primera file del DF
df.drop(0, inplace=True)
df.head()
```

```python
#Renombrar las columnas
columnas = {
    'Tipo de Entidad' : 'TIPO',
    'Entidad Autorizada' : 'ENTIDAD',
    'Compra' : 'COMPRA',
    'Venta' : 'VENTA',
    'Diferencial Cambiario' : 'DIFERENCIAL',
    'ltima Actualizaci贸n' : 'FECHA'

}
df.rename(columns= columnas, inplace=True)
df.head(3)
```

```python
#Proceso transformaci贸n de datos
# Se debe trabajar en el tipado de ls columnas de df
df.info()
columnas_numericas = ['COMPRA', 'VENTA', 'DIFERENCIAL']
df[columnas_numericas] = df[columnas_numericas].apply(pd.to_numeric, errors='coerce')


#df['FECHA'] = df['FECHA'].astype('datetime64[ns]')
df['FECHA'] = df['FECHA'].apply(pd.to_datetime,dayfirst=True, errors='coerce')


df.info()
df.head(3)
```

```python
#Transformaci贸n limpieza de nulos
#Eliminar los valores nulos de df
df.dropna() # Previsualizaci贸n de como quedan los datos
```

```python
#Rellenar datos nulos
df.fillna('NO VALOR').tail()
```

```python
## ffill -> Reemplaza nulos NaN con el 煤ltimo valor de la columna
#inplace= True -> Le indica a el interprete que loca aplicado se guarde en memoria
df.ffill(inplace=True)  # df = df.ffill()
df.tail()
```

```python
#Eliminar la ultima serie del df
#df.drop(37)
df.drop(df.tail(1).index)
df.drop(df.index[-1])

df = df.iloc[:-1] # Seleccionar todas las fila exepto la 煤ltima.
df.tail()
```
